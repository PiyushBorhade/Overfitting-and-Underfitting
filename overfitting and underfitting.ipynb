```python
import pandas as pd
```

```python
df = pd.read_csv('fruit_dataset.csv')
df
```

Checking the shape of data

```python
df.shape
```

Defining X and y

```python
X = df.iloc[:,2:5].values
y = df.iloc[:,0].values
```

Cross Validation

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11)

print('Shape of X_train = ', X_train.shape)
print('Shape of y_train = ', y_train.shape)
print('Shape of X_test = ', X_test.shape)
print('Shape of y_test = ', y_test.shape)
```

Training the data!

```python
from sklearn import tree 
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
print(y_pred)
```

```python
y_pred_train = clf.predict(X_train)
```

```python
print(y_pred_train)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
print("Accuracy on test data is:",accuracy_score(y_test,y_pred))
```

```python
print("Accuracy on training data is:",accuracy_score(y_train,y_pred_train))
```

**We can clearly observe that we are getting 69.23% of accuracy on testing data and we get 100% of accuracy on training data.**

**This is problem of Overfitting!**
